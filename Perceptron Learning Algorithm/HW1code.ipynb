# 0.1.1 (exercise 1.1)

(a) Medical diagnosis:
f:X=>Y is an ideal medical diagnosis algorithm
X: set of possible symptoms
Y: set of possible diseases with given symptoms X

(b) Handwritten digit recognition
f:X=>Y is a formula perfectly matching handwritten digits to (0,9)
X: set of black and white pixels denoting handwritten digits
Y: Digits 0-9

(c) Spam or not?
f:X=>Y is a formula matching features of an email to a spam/nonspam verdict
X: set of (all) possible emails
Y: Just a yes/no verdict

(d) Electric load
f:X=>Y is an perfect function f(price, temperature, day of the week) which returns amount of electric load as an output
X: vector of possible values of price, temperature, day of the week
Y: precise amount of electric load

(e) Analytic vs empirical solution
X: some input data
Y: some output data; often you cannot really explain why X->Y
f:X=>Y is an emulation of an analytic solution using input data. In fact, that is the main goal of machine learning
